## Подготовка файлов для разметки 

Датасет https://www.kaggle.com/datasets/jiashenliu/515k-hotel-reviews-data-in-europe

Этот набор данных содержит 515 000 отзывов клиентов и оценку 1493 отелей по всей Европе.

```python
hotels = pd.read_csv('Hotel_Reviews.csv')
hotels
```

Выделим сэмпл

```python
reviews_sample = hotels.sample(n = 35, replace = False, random_state=42)
```

```python
reviews_sample.shape
```

```python
reviews_sample.head(3)
```

Создадим датасет из колонок Negative_Review и Positive_Review

```python
sample_neg = reviews_sample[["Negative_Review"]]
sample_neg['is_good_review'] = 0
sample_neg = sample_neg.rename(columns = {"Negative_Review":"Review"})
sample_neg.head(5)
```

```python
sample_pos = reviews_sample[["Positive_Review"]]
sample_pos['is_good_review'] = 1
sample_pos = sample_pos.rename(columns = {"Positive_Review":"Review"})
sample_pos.head(5)
```

```python
sample = pd.concat([sample_neg, sample_pos])
sample.head(5)
```

Очистим данные

```python
sample = sample[~sample.Review.isin(['No Negative', 'No Positive'])]
sample.head(5)
```

```python
sample.shape
```

Подготовим пересекающиеся сэмплы для двух людей (все 35 отзывов будут пересекаться).

```python
sample1 = sample.sample(n = 35, replace = False, random_state=42)
sample2 = sample.sample(n = 35, replace = False, random_state=42)
```

Выгрузим их в эксель и отдадим в работу

```python
sample1[['Review']].to_excel('sample1.xlsx')
sample2[['Review']].to_excel('sample2.xlsx')
```
